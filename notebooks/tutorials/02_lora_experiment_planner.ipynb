# LoRA Experiment Planner

Create a reproducible grid for LoRA/QLoRA experiments and emit run cards.

```python
from __future__ import annotations

from dataclasses import dataclass, asdict
from itertools import product
import json
```

```python
@dataclass(frozen=True)
class RunCard:
    base_model: str
    method: str
    lora_r: int
    lora_alpha: int
    learning_rate: float
    batch_size: int
    epochs: int

base_models = ["llama-3-1-8b-instruct"]
methods = ["lora", "qlora"]
lora_r_values = [8, 16]
lora_alpha_values = [16, 32]
learning_rates = [1e-4, 2e-4]
batch_sizes = [2, 4]
epochs = [1, 2]
```

```python
run_cards = [
    RunCard(*params)
    for params in product(
        base_models,
        methods,
        lora_r_values,
        lora_alpha_values,
        learning_rates,
        batch_sizes,
        epochs,
    )
]

len(run_cards)
```

```python
preview = [asdict(card) for card in run_cards[:5]]
print(json.dumps(preview, indent=2))
```
