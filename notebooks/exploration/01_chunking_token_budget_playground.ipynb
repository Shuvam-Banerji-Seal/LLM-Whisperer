# Chunking and Token Budget Playground

Explore simple chunk-size and overlap effects using rough token estimates.

```python
from __future__ import annotations

sample_text = (
    "Agentic systems benefit from clear workflow boundaries, explicit tool policy, "
    "and objective evaluation criteria. Reproducible experiments require stable configs, "
    "versioned datasets, and deterministic quality gates."
)
```

```python
def estimate_tokens(text: str) -> int:
    # Rough heuristic for quick exploration only.
    return max(1, int(len(text.split()) * 1.3))

def chunk_text(text: str, chunk_words: int, overlap_words: int):
    words = text.split()
    chunks = []
    start = 0
    while start < len(words):
        end = min(len(words), start + chunk_words)
        chunks.append(" ".join(words[start:end]))
        if end == len(words):
            break
        start = max(start + 1, end - overlap_words)
    return chunks
```

```python
for chunk_words, overlap_words in [(12, 2), (16, 4), (24, 6)]:
    chunks = chunk_text(sample_text, chunk_words, overlap_words)
    token_estimates = [estimate_tokens(c) for c in chunks]
    print({
        "chunk_words": chunk_words,
        "overlap_words": overlap_words,
        "num_chunks": len(chunks),
        "max_estimated_tokens": max(token_estimates),
    })
```
